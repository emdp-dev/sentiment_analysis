#### Preprocessing
We remove stopwords here. Non alphabetical words are also removed

```python
## add words to stopwords
stopwords = nltk.corpus.stopwords.words('indonesian')
new_stopwords = ['yg','aja','nya','si','dgn','d','gk','tdk','sih','yg','tp','ya','om','bahar','habib','gak','orang','ga','org','dg','lg','bang','Smith','semoga']
stopwords.extend(new_stopwords)
```

```python
stopwords
```

```python
def stopwords_cleaner(text):
  content = [w for w in text if w.lower() not in stopwords and w.isalpha()]
  return content
```

```python
clean_text = stopwords_cleaner(text)
```

### Lexical Diversity
Lexical diversity shows us how often a word appears on a text. It shows us the range of vocab in a given text. 

```python
def lexical_diversity(text):
  return len(set(text)) / len(text)
```

```python
lexical_diversity(clean_text_formatted)
```

###Word Cloud

### Bigrams
PLots words frequently used together on a bar graph. Collocations is better

```python
from nltk.collocations import *
from nltk.util import ngrams
```

```python
bigrams = ngrams(clean_text,2) #select number of n grams
```

```python
bigrams_count = Counter(bigrams) # count number of 
```

```python
bigram_threshold = 70 #remove below 4
```

```python
bigrams_count = {x: count for x, count in bigrams_count.items() if count >= bigram_threshold} # get count => 4
```

```python
sorted_bigramscount = sorted(bigrams_count)
```

```python
sorted_bigramscount
```

```python
sorted_bigramscount
```

```python
names = list(bigrams_count.keys())   #make this sorted later
actual_name = []

for n in names:
  actual_name.append(n[0] + '_' +n[1])

values = list(bigrams_count.values())
plt.figure(figsize=(20,10))
plt.bar(range(len(bigrams_count)), values, tick_label=actual_name)
```

### Trigrams

```python
trigrams = ngrams(clean_text,3) #select number of n grams
trigrams_count = Counter(trigrams) # count number of 
trigrams_threshold = 100 #remove below 4
trigrams_count = {x: count for x, count in trigrams_count.items() if count >= trigrams_threshold} # get count => 4

names = list(trigrams_count.keys())   #make this sorted later
actual_name = []

for n in names:
  actual_name.append(n[0] + '_' +n[1])

values = list(trigrams_count.values())
plt.figure(figsize=(15,8))
plt.bar(range(len(trigrams_count)), values, tick_label=actual_name)
```
