```python
for tweet in tweets_copy:
    hashtags = []
    mention = []
    try:
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
        author = api.get_status(id=tweet.id, tweet_mode='extended').retweeted_status.author.screen_name

        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        for mention in tweet.entities["user_mentions"]:
            mention.append(tweet.entities["user_mentions"][0].get('screen_name'))
        
    except:
        pass
    
    tweets_df = tweets_df.append(pd.DataFrame({'user_name': tweet.user.screen_name,
                                               'nama' : tweet.user.name,
                                               'id' : tweet.user.id,
                                               'retweet_count' : tweet.retweet_count,
                                               'author' : author,
                                               'mention' :[mention['screen_name'] if mention else None],
                                               'date': tweet.created_at,
                                               'text': text, 
                                               'hashtags': [hashtags if hashtags else None],
                                               'source': tweet.source}))
    tweets_df = tweets_df.reset_index(drop=True)
```
